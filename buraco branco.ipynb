```python
class buraco_branco():
    construtor = "god_constructs"

    def __init__(self, galaxia, sistema_solar, cordenadas):
       self.galaxia = galaxia
       self.sistema_solar = sistema_solar
       self.cordenadas = cordenadas

    def getconstrutor(self):
        return self.construtor

    def getgalaxia(self):
        return self.galaxia

    def getsistema_solar(self):
        return self.sistema_solar

    def getcordenadas(self):
        return self.cordenadas

    def setconstrutor(self, construtor):
        self.construtor = construtor

    def setgalaxia(self, galaxia):
        self.galaxia = galaxia

    def setsistema_solar(self, sistema_solar):
        self.sistema_solar = sistema_solar

    def setcordenadas(self, cordenadas):
        self.cordenadas = cordenadas


    def enderecocompleto(self):
       return (f"endereço completo: galaxia:{self.galaxia} sistema solar:{self.sistema_solar} cordenadas:{self.cordenadas} ")
```

```python
buraco_branco1 = buraco_branco(galaxia="via lactea", sistema_solar="estrela Drun", cordenadas=[-1122343432, 3212312344, 9020110])
```
